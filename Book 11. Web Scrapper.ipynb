<h1>Book 11. Web Scrapper</h1><hr>

<b>Web scraping</b> is a term used to describe the use of a program or algorithm to extract and process large amounts of data from the web. Whether you are a data scientist, engineer, or anybody who analyzes large amounts of datasets, the ability to scrape data from the web is a useful skill to have. Let's say you find data from the web, and there is no direct way to download it, web scraping using Python is a skill you can use to extract the data into a useful form that can be imported.

Web scrapper is a program that can scrap data from a website. By scaning the HTML elements in the website. 

We will be using <b>BeautifulSoup</b> and <b>requests</b> package to do it.

```python
from bs4 import BeautifulSoup

import requests
```

Now lets pass a request to a website to get data from it.

```python
r  = requests.get("http://boingboing.net")
print(r)
```

Response code <b>200</b> indicates the connection was successful

<b>Connection Status</b><br>
<ul>
    <li><u>1xx (information)</u>: The server received the request but some extra information for the clinet.
    </li>
      <li><u>2xx (success)</u>: It worked: every success code other than 200 conveys extra details.
    </li>
      <li><u>3xx (redirection)</u>: The resource moved, so the response returns the new URL to the client.
    </li>
      <li><u>4xx (client error)</u>: Some problem from the client side, such as famous 404(not found). 418 (I'm a teapot) was an April Fool's joke.
    </li>
      <li><u>5xx (server error)</u>: 500 is the generic whoops; you might see a 502(bad gateway) if there's some disconnect between a web server and a backend application server.
    </li>
</ul>

```python
data = r.text
print(data)
# r.text gives us the text content got by the request
# below what you can see is the text content of the request
# content of request is always got as HTML format. This is the same format which rendered on your browser. And browser renders HTML to a webpage UI.
```

```python
soup = BeautifulSoup(data,'lxml')
print(soup)
# BeautifulSoup(data,'lxml') is setting up a parser to scan the data
```

```python
# Lets scan for a particular thing. Lets scan for all links in the webpage.
# Links are coded as anchor tag(<a href="link">). Hence lets find all the anchor tag and get the link which it is pointing too. 
for link in soup.find_all('a'): # scarping the links by extracting the text in href of <a> tag(HTML) 
    print(link.get('href')) 
```

<span style='color:green'>Tadaa, we were able to compete our objective and extract the links. There might be chances of few non-links also to appear.</span>

<span style='color:red'>Note: Web technology is evolving day by day and lot of frameworks are advancing. Most of them render content dynamically and hence the content cannot be caught in response text</span>
